```python
def extract_features(signal):
    return [       
        librosa.feature.zero_crossing_rate(signal).mean(),
        librosa.feature.mfcc(signal).mean(),
    ]
```

```python
#Compute features
signal_features = np.array([extract_features(x) for x in audio_signals])
signal_features
```

**Classificador K-means**

```python
#normalizar features
scaler = sklearn.preprocessing.MinMaxScaler(feature_range=(-1, 1))
training_features = scaler.fit_transform(signal_features)
print(training_features.min(axis=0))
print(training_features.max(axis=0))
```

```python
plt.scatter(training_features[0:5,0], training_features[0:5,1], c='b')
plt.scatter(training_features[5:10,0], training_features[5:10,1], c='r')
plt.scatter(training_features[10:,0], training_features[10:,1], c='g')
plt.xlabel('Zero Crossing Rate')
plt.ylabel('MFCC')
plt.legend(('dog_bark', 'siren', 'engine'))
```

```python
#labels_true = np.concatenate((np.zeros(5), np.ones(5), -1*np.ones(5)))
#labels_true
```

```python
#Apply k-means
model = sklearn.cluster.KMeans(n_clusters=3)
labels_pred = model.fit_predict(training_features)
print(labels_pred)
```

```python
#plotar resultado
plt.scatter(training_features[labels_pred==0,0], training_features[labels_pred==0,1], c='b')
plt.scatter(training_features[labels_pred==1,0], training_features[labels_pred==1,1], c='r')
plt.scatter(training_features[labels_pred==2,0], training_features[labels_pred==2,1], c='g')
plt.xlabel('Zero Crossing Rate')
plt.ylabel('MFCC')
plt.legend(('cluster1', 'cluster2', 'cluster3'))
```
